countour, image and persp

```r
x = seq(-pi, pi, length = 10)
y = x

z = outer(x, y, function(x, y) cos(y)/(1 + x^2))
```

```r
contour(x,y,z)
```

```r
image(x,y,z)
```

```r
persp(x, y, z, theta = 30, phi = 20)
```
