### Задача 4 (a)(b)

Меня заинтерисовала задача 4 которая предлагает сравнить результаты линейной регрессии и полинома 3-й степени на выборке с реальной линейной взаимосвязью.<br>
Генерация обучающей и тестовой выборок.

```r
n = 100
set.seed(20)

train.df = data.frame(x = sort(runif(n, 0, 40)))
train.df["y"] = train.df["x"]*4 + 2 + runif(n, -20,20)

test.df = data.frame(x = sort(runif(n, 0, 40)))
test.df["y"] = test.df["x"]*4 + 2 + runif(n, -20,20)
```

```r
describe_y_hat = function(y_hat, df)
{
    RSS = sum((y_hat - df$y)^2)
    print(paste("RSS =", as.character(RSS)))

    plot(df$x, df$y)
    lines(df$x, y_hat)
}
```

**Линейная модель**

```r
linear.model = lm(y ~ x, data = train.df)
summary(linear.model)
```

Обучающая выборка

```r
y_hat = predict(linear.model, train.df)
describe_y_hat(y_hat, train.df)
```

Тестовая выборка

```r
y_hat = predict(linear.model, test.df)
describe_y_hat(y_hat, test.df)
```

**Кубическая модель**

```r
cube.model = lm(y ~ poly(x, 3), data = train.df)
summary(cube.model)
```

```r
y_hat = predict(cube.model)
describe_y_hat(y_hat, train.df)
```

```r
y_hat = predict(cube.model, test.df)
describe_y_hat(y_hat, test.df)
```

### Задача 7

Не выходит доказать - упираюсь в тупик. Сделаю данные удовлетворяющие условию задачи и буду подставлять в разные места тождества пока не найду ошибку.

```r
n = 200
set.seed(4324)

x = runif(n, -20, 20)
y = x*9 + runif(n, -50,50)

x = x - mean(x)
y = y - mean(y)
```

```r
model = lm(y ~ x)
summary(model)

y_hat = predict(model)
beta_1 = summary(model)$coefficients[2]
```

Коэффициент корреляции

```r
cor(x, y)
```

Коэффициент корреляции по формуле $$r = \frac{\sum_{i=1}^n{x_iy_i}}{\sqrt{\sum_{i=1}^n{x_i^2} \sum_{i=1}^n{y_i^2}}}$$

```r
(sum(x*y))/sqrt(sum(x^2)*sum(y^2))
```

Коэффициент дескриминации по формуле

$$R^2 = \frac{\sum{y_i^2} - \sum{(y_i - \hat{y}_i)^2}}{\sum_{i=1}^n{y_i^2}}$$

```r
(sum(y^2) - sum((y-y_hat)^2))/sum(y^2)
```

Коэффициент уравнения регрессии

$$\hat{\beta}_1 = \frac{\sum_{i=1}^nx_iy_i}{\sum_{i=1}^n{x_i^2}}$$

```r
sum(x*y)/sum(x^2)
```

Выполняется ли доказываемая теорема?

$$R^2 = r^2$$

```r
cor(x,y)^2
```

Задача сводилась к выражению

$$\sum_{i=1}^n{y_i^2} - \sum_{i=1}^n{(y_i - \hat{y}_i)^2} = \frac{(\sum_{i=1}^n{x_iy_i})^2}{\sum{x_i^2}}$$


```r
sum(y^2) - sum((y-y_hat)^2)
```

```r
sum(x*y)^2/sum(x^2)
```

Преобрузя левую часть можно получить, что
$$\sum_{i=1}^n{2y_i\hat{y}_i} - \sum_{i=1}^n{\hat{y}^2_i} = \frac{(\sum_{i=1}^n{x_iy_i})^2}{\sum{x_i^2}}$$

```r
sum(2*y*y_hat) - sum(y_hat^2)
```

Подаставляя уравнение регрессии в выражение получим

$$\sum_{i=1}^n{[2y_i\hat{\beta}_1x_i - (\hat{\beta_1}x_i)^2]} = \frac{(\sum_{i=1}^n{x_iy_i})^2}{\sum{x_i^2}}$$

```r
sum(2*y*beta_1*x - (beta_1*x)^2)
```

Подставляя выражение для коэффициента можно получить что

$$\frac{(\sum_{i=1}^n{x_iy_i})^2}{\sum{x_i^2}}\sum_{i=1}^n{[2y_ix_i + \frac{x_i^2}{\sum_{j=1}^n{x_j^2}}]} = \frac{(\sum_{i=1}^n{x_iy_i})^2}{\sum{x_i^2}}$$

Для того чтобы это выполнялось надо чтобы 

$$\sum_{i=1}^n[{\frac{2y_ix_i}{\sum{x_jy_j}} - \frac{x_i^2}{\sum_{j=1}^n{x_j^2}}]} = 1$$

```r
sum(2*y*x/sum(x*y) - x^2/sum(x^2))
```

На ст. 112 сказано, что среднее значение показателя разбалансировки должно равнятся $\frac{p+1}{n}$ - убедимся в этом.

Но для полной диагностики следует еще изучить стьюдентизированные остатки.

```r
r_resid = rstudent(model)
plot(x = h, y = r_resid)
```

Попробуем сформировать критерий отбора точек, на графике выше - проведем две прямые которые отделят хорошие наблюдения от плохих. И будем пробовать выбрасывать плохие наблюдения смотреть как это повлияет на качество модели.

```r
h.range = seq(0, 0.06, length = 20)

get_line = function(beta_0, beta_1, range)
{
    # Получить занчение линейной функции при заданных коэффиентах для заданного
    # массива данных
    
    return (beta_0 - beta_1*range)
    
}

get_line(2, 60, h.range)
```

```r
resid.range = get_line(2, 20, h.range)

plot(x = h, y = r_resid)
lines(x = h.range, y = resid.range)
lines(x = h.range, y = -resid.range)

summary(lm(
    Sales ~ Price + US, 
    data = Carseats[(resid.range > r_resid) & (-resid.range < r_resid) ,]
))
```

```r
resid.range = get_line(2, 50, h.range)

plot(x = h, y = r_resid)
lines(x = h.range, y = resid.range)
lines(x = h.range, y = -resid.range)

summary(lm(
    Sales ~ Price + US, 
    data = Carseats[(resid.range > r_resid) & (-resid.range < r_resid) ,]
))
```

```r
resid.range = get_line(1.5, 20, h.range)

plot(x = h, y = r_resid)
lines(x = h.range, y = resid.range)
lines(x = h.range, y = -resid.range)

summary(lm(
    Sales ~ Price + US, 
    data = Carseats[(resid.range > r_resid) & (-resid.range < r_resid) ,]
))
```

По результатам исследования как и сказано в книге на ст. 111 выбросы сильнее влияют на $R^2$ модели чем разбалансированные значения
